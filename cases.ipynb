```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/joshuagrajales/Desktop/qac211/project/SM_comm/zoom/all.csv")
fig, axs1 = plt.subplots(1, 2, figsize = (20,10))
d20 = df['Date'][df['Date'] <= '2020-10-28'][df['Date'] >= '2020-10-01']
st20 = df['All Stock'][df['Date'] <= '2020-10-28'][df['Date'] >= '2020-10-01']
se20 = df['Polarity'][df['Date'] <= '2020-10-28'][df['Date'] >= '2020-10-01']

axs1[0].scatter(d20,st20, color = 'blue')
axs1[0].plot(d20,st20, color = 'blue')
axs1[0].grid()
axs1[0].set_title("Stocks Over Time")
axs1[0].set_xlabel("Date")
axs1[0].set_ylabel("Closing Stock Price ($)")

axs1[1].scatter(d20,se20, color = 'blue')
axs1[1].plot(d20,se20, color = 'blue')
axs1[1].grid()
axs1[1].set_title("Sentiment Over Time")
axs1[1].set_xlabel("Date")
axs1[1].set_ylabel("Polarity")
x = input("Title of graph: ")
fig.suptitle(x, fontsize = 15)
fig.autofmt_xdate()
```

```python
zd_d21 = df['Date']["2021-12-01"<=df['Date']]["2021-12-30" >= df['Date']]
zd_st21 = df['All Stock']["2021-12-01"<=df['Date']]["2021-12-30" >= df['Date']]
zd_se21 = df['Polarity']["2021-12-01"<=df['Date']]["2021-12-30" >= df['Date']]
fig2, axs = plt.subplots(1, 2, figsize = (20,10))

axs[0].scatter(zd_d21,zd_st21, color = 'blue')
axs[0].plot(zd_d21,zd_st21, color = 'blue')
axs[0].grid()
axs[0].set_title("Stocks Over Time")
axs[0].set_xlabel("Date")
axs[0].set_ylabel("Closing Stock Price ($)")

axs[1].scatter(zd_d21,zd_se21, color = 'blue')
axs[1].plot(zd_d21,zd_se21, color = 'blue')
axs[1].grid()
axs[1].set_title("Sentiment Over Time")
axs[1].set_xlabel("Date")
axs[1].set_ylabel("Polarity")
x = input("Title of graph: ")
fig2.suptitle(x, fontsize = 15)
fig2.autofmt_xdate()
```
